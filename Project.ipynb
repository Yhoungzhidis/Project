> CHECKING FOR MISSING VALUES IN THE DATA SET

```python
dataSet.isnull().sum()
```

```python
median_fa = Stats.fixed_acidity.loc['50%']
median_va = Stats.volatile_acidity.loc['50%']
median_cl = Stats.chlorides.loc['50%']
median_fs = Stats.free_sulfur_dioxide.loc['50%']
median_de = Stats.density.loc['50%']
median_qu = Stats.quality.loc['50%']
```

```python
new_dataSet = dataSet.copy() 
```

```python
new_dataSet.fixed_acidity       = dataSet.fixed_acidity.fillna(median_fa)
new_dataSet.volatile_acidity    = dataSet.volatile_acidity.fillna(median_va)
new_dataSet.chlorides           = dataSet.chlorides.fillna(median_cl)
new_dataSet.free_sulfur_dioxide = dataSet.free_sulfur_dioxide.fillna(median_fs)
new_dataSet.density             = dataSet.density.fillna(median_de)
new_dataSet.quality             = dataSet.quality.fillna(median_qu)
```

## GRAPHS PLOTTING

> ## HISTOGRAM

NB: This is based on the density

```python
bins = np.arange(0.990, 1.004,0.002)

mat.figure(figsize = (8,7))
mat.hist(new_dataSet.density, bins = bins, rwidth = 0.8)
mat.xticks(bins)

mat.title('DENSITY OF THE RED WINES', fontdict = {'fontweight' : 'bold'})
mat.xlabel('Density', fontsize = 12, fontstyle = 'italic')
mat.ylabel('Red Wine', fontsize = 12, fontstyle = 'italic')

mat.show()
```

From the histogram, it can be seen that most red wines had their densities between 0.994 and 1.000. On the other hand, few red wines had their densities between 0.990 and 0.994 and between 1.000 and 1.004. Also, it was gotten from the new Dataset table.

> ## SCATTERED GRAPH

NB: This graph is based on fixed acidity against pH.

```python
mat.figure(figsize = (8,7))
mat.title('FIXED ACIDITY AGAINST PH', fontdict = {'fontweight' : 'bold'})
mat.scatter(x = 'pH', y = 'fixed_acidity', data = new_dataSet, alpha = 0.8, c = 'r')

mat.xlabel('pH', fontsize = 12, fontstyle = 'italic')
mat.ylabel('Fixed Acidity', fontsize = 12, fontstyle = 'italic')

mat.legend()
```

From the scatter plot depicting the relationship between the fixed acidity and pH of red wines, it is evident that those with lower pH levels exhibit higher fixed acidity, thus creating a discernible negative slope. In simplier terms, as the pH of red wine increases, its fixed acidity tends to decrease. This implies an inverse correlation between the pH and fixed acidity in the dataset. Below is the table being used for the graph.

```python
nData = new_dataSet.filter(items = ['fixed_acidity', 'pH'])
```

> ## PIE CHART

NB: This is based on the average amount of residual sugar, sulphates, alcohol, free sulphur dioxide, citric acid and chlorides from the dataset.

```python
Stats_2[['residual_sugar', 'sulphates', 'alcohol', 'free_sulfur_dioxide', 'citric_acid', 'chlorides']].loc['mean']
```

```python
ResidualSugar     = Stats_2['residual_sugar'].loc['mean']
Sulphates         = Stats_2['sulphates'].loc['mean']
Alcohol           = Stats_2['alcohol'].loc['mean']
FreeSulfurDioxide = Stats_2['free_sulfur_dioxide'].loc['mean']
CitricAcid        = Stats_2['citric_acid'].loc['mean']
Chlorides         = Stats_2['chlorides'].loc['mean']

labels = ['Residual sugar', 'Sulphates', 'Alcohol', 'Free sulfur dioxide', 'Citric acid', 'Chlorides']
Values = [0,0,0,0,0,0.6]

mat.pie([ResidualSugar, Sulphates, Alcohol, FreeSulfurDioxide, CitricAcid, Chlorides],
labels = labels, autopct = '%.2f %%', explode = Values)
mat.title('THE COMPONENT OF THE RED WINE', fontdict = {'fontweight' : 'bold'}, ha = 'center')
mat.show()
```

While the detail about the individual qualities of various red wines were given, the analysis employed the mean quality of each component. This approach was taken to streamline the comparison, treating the components as if they belong to a single red wine. The goal was to assess the percentage composition in red wine based on this aggregated perspective.
